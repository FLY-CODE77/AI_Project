```python
import cv2
import os, glob, numpy as np


# categories 
categories = []
for i in range(1000):
    categories.append(str(i))

nb_classes = len(categories)

# load img
caltech_dir = "/mnt/nvme1n1/LG_CNN/train"

X = []
y = []

for idx, file in enumerate(categories):
    label = [0 for i in range(nb_classes)]
    label[idx] = 1
    
    image_dir = caltech_dir + "/" + file
    files = glob.glob(image_dir + "/*.jpg")
    print(file, " 파일 길이 : ", len(files))
    for i, f in enumerate(files):
        img = cv2.imread(f,cv2.IMREAD_COLOR)
        img = cv2.normalize(img, None, 0, 1, cv2.NORM_MINMAX)
        X.append(img)
        y.append(label)
        
```

```python
import pickle

with open('X.pickle', 'wb') as f:
    pickle.dump(X, f)
    
with open('y.pickle', 'wb') as f:
    pickle.dump(y, f, protocol=4)
```

```python
import pickle

with open('X.pickle', 'rb') as f:
    X = pickle.load(f,encoding = 'utf-8')

with open('y.pickle', 'rb') as f:
    y = pickle.load(f, encoding = 'utf-8')
```

```python
import numpy as np
X_train = np.array(X)
y_train = np.array(y)
del X,y
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, train_size = 0.7)
# del X,y
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), padding="same", input_shape=X_train.shape[1:], activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3,3), padding="same", activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_dir = './model'

if not os.path.exists(model_dir):
    os.mkdir(model_dir)

model_path = model_dir + '/multi_img_classification.model'
checkpoint = ModelCheckpoint(filepath=model_path , monitor='val_loss', verbose=1, save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=50, validation_split=0.2, callbacks=[checkpoint, early_stopping])
```
